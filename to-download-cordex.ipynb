```python
import requests
import xarray
import pathlib

wget_paths = list(pathlib.Path().glob('*.sh'))


##wget_script = 'mohc-pr-rcp85.txt'

def extract_files_info(wget_script):
    start = 'download_files="$(cat <<EOF--dataset.file.url.chksum_type.chksum'
    end = 'EOF--dataset.file.url.chksum_type.chksum'

    with open(wget_script,'r') as f:
        inside = False
        files_info_dictionary = []
        for line in f:
            if start in line:
                inside = True
                continue
            elif end in line:
                inside = False
                break
            
            if inside:
                #print(line, end='')
                parts = [part.strip("'") for part in line.split()]
                file_name,file_url,idk1,idk2 = parts
                #print(f"File name: {file_name}")
                #print(f'File url:{file_url}')
                temp_dict = {'name':file_name,'url':file_url}
                files_info_dictionary.append(temp_dict)
    return files_info_dictionary

def download_file(file_info):
    file_name = file_info['name']
    file_url = file_info['url']

    response = requests.get(file_url,stream=True)
    if response.status_code == 200:
        with open(file_name,'wb') as f:
            for chunk in response.iter_content(1024):
                f.write(chunk)

    else:
        print(f"Failed to download {file_name}")

def batch_download(wget_script):
    files_info_dictionary = extract_files_info(wget_script)
    for file_info in files_info_dictionary:
        print(f' Downloading [{file_info["name"]}]...')
        download_file(file_info)
        print(f'Download  completed for [{file_info["name"]}]')

for wget_script in wget_paths:
    batch_download(wget_script)
    
        
```
